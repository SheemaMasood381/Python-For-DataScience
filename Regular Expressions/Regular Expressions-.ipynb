```python
dates ="""1. Today is 02/13/2024.
2. The event is scheduled for 2024-02-14.
3. The deadline for submissions is 15-February-2024.
4. Our anniversary is on 2024/02/16.
5. Please mark your calendars: 02-17-2024.
6. The meeting is on March 18th, 2024.
7. The conference starts on 2024/03/19 and ends on 03/21/2024.
8. Don't forget about the appointment on April 22, 2024.
9. The training session is scheduled for 04/23/2024.
10. Remember to submit the report by May 24, 2024."""
```

```python
patterns =r'\b\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4}|[A-Z][a-z]+\s+\d{1,2}th,\s+\d{4}'
datesregex = re.findall(patterns , dates)
```
