# How Many Appointments in data set

## Particular Date & Time

```python
data_list["appt_start_date"].unique()
```

```python
data_list["appt_start_date"].nunique()
```

```python
appointments_by_date = data_list.groupby("appt_start_date").size().reset_index(name="num_appointments")
print(appointments_by_date)
```

```python
data_list["appt_start_date"].value_counts().sort_values(ascending=False) #sort_index()
```

## Particular Date

```python
import datetime as dt
```

```python
date_format = "%m/%d/%y %H:%M"
```

```python
data_list.head()
```

```python
asd = pd.to_datetime(data_list["appt_start_date"], format=date_format)
#print(asd)
dis = asd.dt.normalize() # use to display without time
dis.value_counts().sort_values(ascending=False) #sort_index()
```

```python
pd.to_datetime(data_list["appt_start_date"], format=date_format).dt.normalize()
```
