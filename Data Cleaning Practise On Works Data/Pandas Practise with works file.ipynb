# fx for clean date

```python
def clean_date(date):
    bad_char = ["(",")","C","c", ".", " ","s", "'"]
    for i in bad_char:
        date = date.replace(i,"")
    return date
```

```python
clean_date("149s")
```

```python
df["Date"].apply(clean_date)
```

```python
df["Date"] = df["Date"].apply(clean_date)
```

```python
def int_date(date):
    if "-" in date: # membership operator
        lst = new_date.split("-")
        yr1 = int(lst[0])
        yr2 = int(lst[1])
        date = round((yr1+yr2)/2)
    else:
        date = int(date)
    return date
```
