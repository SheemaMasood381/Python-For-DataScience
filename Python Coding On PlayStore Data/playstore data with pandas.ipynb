# creating function to identify apps with engliash names

```python
def eng_app(word_name):
    non_ascii = 0
    for i in word_name:
        word_ordi = ord(i)
        if word_ordi > 127:
            non_ascii += 1
    if non_ascii > 3:
        return False
    else:
        return True
```

```python
#creating bool for english apps and saving it in variable
bol_eng_app = data_df["App"].apply(eng_app)
bol_eng_app.sum()       #only true values
```
